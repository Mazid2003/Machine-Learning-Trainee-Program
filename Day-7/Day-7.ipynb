Example


1. Loading sample dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic=sns.load_dataset('titanic')
tips=sns.load_dataset('tips')
```

2. Distribution Plots

```python
# Plotting Histogram
sns.histplot(data=tips, x='total_bill', kde=True)
plt.title("Distribution of Total Bill")
plt.show()
```

```python
# Plotting kdeplot()  Kernel Density Estimate
sns.kdeplot(data=tips['total_bill'], fill=True, color='green')
```

```python
# boxplot()  Box and Whisker Plot
sns.boxplot(data=tips, x='day', y='total_bill', palette='Set2')
```

```python
# violinplot()  Combines boxplot and KDE
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)
```

3. Relationship Plots

```python
# scatterplot()
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
# lineplot()
sns.lineplot(data=tips, x='size', y='tip', estimator='mean')
```

```python
# jointplot()  2D Distribution
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
```

```python
# pairplot()  Matrix of scatter plots
sns.pairplot(tips, hue='sex')
```

4. Categorical Plots


```python
# countplot()
sns.countplot(data=titanic, x='class', hue='sex')
```

```python
# barplot()  With confidence intervals
sns.barplot(data=tips, x='day', y='total_bill', hue='sex', ci='sd')
```

```python
# stripplot()  Categorical scatter plot
sns.stripplot(data=tips, x='day', y='total_bill', jitter=True)
```

 5. Styling and Themes


```python
# Set Theme
sns.set_theme(style='darkgrid')  # other options: white, whitegrid, dark, ticks
# Custom Color Palettes
sns.set_palette('pastel')  # or try 'deep', 'muted', 'bright', 'dark'
```

6. Advanced Visualization Use Cases

```python
# Titanic Dataset  Survival by Gender and Class
sns.catplot(data=titanic, x='sex', hue='survived', col='class', kind='count')
```

```python
# Correlation Heatmap
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
